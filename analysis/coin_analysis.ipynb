```python
## 8) Fiyat Normalizasyonu ve Göreceli Performans (Benchmarking)

# Farklı fiyatlardaki coinleri karşılaştırmak için normalizasyon yapalım
# Seçilen bir tarihteki fiyatı "100" kabul ederek endeks grafiği oluşturalım

# Örnek: İlk tarih (en eski ortak tarih) baz alınarak normalize
if not df_prices.empty:
    # İlk tarih için fiyatları 100'e normalize et
    df_index = df_prices / df_prices.iloc[0] * 100
    df_index.plot(title='Normalize Edilmiş Endeks Grafiği (Başlangıç = 100)', figsize=(12,6))
    plt.ylabel('Endeks Değeri (100 = Başlangıç)')
    plt.grid(True)
    plt.show()

    # Son 30 günün performansını hesapla
    if len(df_index) > 30:
        df_recent = df_index.tail(30)
        performance_30d = (df_recent.iloc[-1] / df_recent.iloc[0] - 1) * 100
        print("Son 30 Gün Performansı (%):")
        print(performance_30d.sort_values(ascending=False))

        # Örnek analiz: En iyi ve en kötü performanslı coinleri belirle
        best_coin = performance_30d.idxmax()
        worst_coin = performance_30d.idxmin()
        print(f"\nEn iyi performans: {best_coin} (%{performance_30d[best_coin]:.1f})")
        print(f"En kötü performans: {worst_coin} (%{performance_30d[worst_coin]:.1f})")

        # Kısa analiz metni
        analysis = f"Son 30 günde {best_coin} %{performance_30d[best_coin]:.1f} artış gösterirken, {worst_coin} %{performance_30d[worst_coin]:.1f} düşüş yaşadı."
        print(f"\nAnaliz: {analysis}")
    else:
        print("Yeterli veri yok (30 günden az)")
else:
    print("df_prices boş, normalizasyon yapılamadı.")
```
